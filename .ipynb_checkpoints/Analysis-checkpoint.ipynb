```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("darkgrid", {"axes.facecolor": ".6"})
```

```python
df=pd.read_csv('avirus.csv',encoding='latin1')
df=df.dropna(subset=['text'])
df
```

```python
from textblob import TextBlob
x=df['text']
senti=[]
for i in x:
    senti.append(TextBlob(i).sentiment[0])
    #print (TextBlob(i).sentiment[0])
df['senscores']=senti    
df
```

```python
df['month']=df['timestamp'].apply(lambda x :x[5:7])
df
```

```python
df['day']=df['timestamp'].apply(lambda x :x[8:10])
df
```

```python
from datetime import datetime

hour=[]
for i in df['timestamp_epochs']:
    ts = int(i)
    #print (datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S'))
    hour.append(datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')[11:13])
print (hour)
df['hour']=hour
df
```

```python
data=df.groupby('month').count()
data
```

```python
x=['November','December','January','February','March','April']

y=[2061, 2032,2801, 2101, 2119, 2807]
```

```python
fig, ax = plt.subplots(figsize=(15,7))
plt.bar(x, y, edgecolor = 'black', linewidth=2)

ax.set_xlabel('Months',fontsize=20)
ax.set_ylabel('Number of tweets',fontsize=20)
ax.set_title("Distribution of total tweets vs months",fontsize=20)

ax = plt.gca()
ax.tick_params(axis = 'both', which = 'major', labelsize = 15)    
    
```

```python
from collections import Counter

di=dict(Counter(df['username']).most_common(20))
di
```

```python
x=[]
y=[]
for i,j in di.items():
    x.append(i)
    y.append(j)
x,y    
```

```python
keys, values = di.keys(), di.values()
print (list(keys),list(values))
```

```python
fig, ax = plt.subplots(figsize=(15,7))
plt.bar(x,y, edgecolor = 'black', linewidth=2)
plt.xticks(rotation=90)
ax.set_xlabel('Usernames',fontsize=20)
ax.set_ylabel('Number of tweets',fontsize=20)
ax.set_title("Users who tweet the most",fontsize=20)

ax = plt.gca()
ax.tick_params(axis = 'both', which = 'major', labelsize = 15)    
    
```
